```python
#Question 5
```

```python
marks = {'Andy':88, 'Amy':66, 'James':55, 'Arthur':77}
print(marks)
```

```python
names = list(marks.keys())
grades = list(marks.values())

plt.bar(range(len(marks)), grades, tick_label=names)
plt.title("Grades")
plt.show()
```

```python
names = list(marks.keys())
grades = list(marks.values())

plt.pie(grades, labels=names)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.title("Grades")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot1
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])

x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])

x3 = np.array([0, 1, 2, 3])
y3 = np.array([5, 10, 40, 30])

plt.subplot(2, 3, 1)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.title("Mult. Lines")

#plot 2
plt.subplot(2, 3, 2)
x = np.array(["A", "B", "C"])
y = np.array([4, 6, 8])
plt.bar(x, y, color = 'm')
plt.title("Grades Bar")

#plot3
plt.subplot(2, 3, 3)
plt.pie(grades, labels=names)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.title("Grades Pie")

#plot 4
plt.subplot(2, 3, 4)
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 20, 40])
plt.plot(x,y)
plt.grid()
plt.title("Grid")

#plot 5
plt.subplot(2, 3, 5)
x = np.array([5,6,7,2,3,4,11,14,15,18])
y = np.array([10,11,15,16,19,23,25,27,32,40])
plt.scatter(x,y)
plt.title("Scatter")

#plot 6
plt.subplot(2, 3, 6)
mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, 1000)
plt.hist(s)
plt.title("Histogram")

plt.tight_layout(pad = 1.0)
plt.show()
```
